Let us start by defining the terms Flexible Statistical Learning and Inflexible Statistical Learning

a. FLEXIBLE STATISTICAL LEARNING

i.	This refers to a machine learning model that is more accommodative to complex data. It can adapt easily and are capable of fitting a variety of feature variable relations with the target variables.

ii.	They are also prone to overfitting due to high flexibility and usually involve a large number of parameters.

iii.	Ex: Decision Trees, Neural Networks etc

iv.	Better for larger and relatively complex datasets. 



b.	INFLEXIBLE STATISTICAL LEARNING

i.	This refers to models that are strict in terms of assumptions of the data. They tend to have limited adaptability to complex relationships in the data. But this inturn also makes them more stable

ii.	They tend to have higher bias and can lead to underfitting and usually involve fewer parameters making them less prone to over fitting

iii.	Ex: Linear Regression, SVM etc

iv.	Better for smaller datasets having clear structures.



Now for our conditions:


a.	Large Sample Size, Less Predictors:
Here, a Flexible method would perform better than an unflexible method largely due to the large sample size. This large data set can prevent the flexible model from overfitting while at the same time capture complex relationships within the dataset. An inflexible model might not be able to capture the patterns in a larger dataset due to its rigidity.

    
b.	Large Predictors, Less observations:
Here, an unflexible method would perform better than a flexible method again due to the sample size. A flexible model can have the tendency to capture the noise within the data set and cause over fitting. An unflexible model is strict enough and works best in datasets with limited data or “simple structures”. This can result in accurate prediction and prevent any sort of overfitting.


	
c.	Highly Non-Linear Relationship between Predictors and Response:
In this case, Flexible method would perform better than Unflexible method because an unflexible method typically tends to assume a linear relationship between data and in a situation where the relationship is non-linear, the method could struggle. Flexible methods are capable to capture this type of relation more effectively.

(e). Increasing the value of K in KNN essentially means we are increasing the number of neighbors considered when predicting a result. Due to this, increasing the value of ‘K’ will result the model becoming relatively less flexible. 

To understand this, we can look at how KNN predicts a value. KNN uses the result of ‘K’ number of neighbors to arrive at a result.

For a low ‘K’ value, the model will only consider a small number of its closest neighbors, which means it can adapt well to any noise or other variations in the dataset. This will result in overfitting where the result will be based off noise rather than the actual dependencies. 
This can lead to a model that will fit the training data very well, but will perform poorly in unseen data. 
Thus, we can say that for a low ‘K’ value, the model will be flexible. 

When we increase the ‘K’ value, we are actually increasing the number of neighbors across which the result is determined. Since we are considering more neighbors, the effect of any isolated localized fluctuations is reduced and the overall average of the large number of alternate data points makes sure that the results are more stable. 
The boundaries will be much smoother across a number of datapoints, as opposed to a lower ‘K’ value where the boundary will be jagged to fit (overfit) all the points within its radius. 
In this case, the model will have a relatively lower variance and higher stability since it relies on a larger number of neighbors but also a higher bias because it might ignore individual fluctuations.
Thus, we can say that for a higher ‘K’ value, the model becomes decreasingly flexible.  


1)	Square of Bias:


i.	For a low ‘K’ value, the bias will be low since the model captures the training data very well and since it is highly sensitive to small fluctuations in its immediate radius. 

ii.	For a higher ‘K’ value, the bias will increase since the model averages across a larger number of neighbors and as a result may ignore local fluctuations. 


2)	Variance:


i.	For a low ‘K’ value, the model is highly sensitive to smaller fluctuations and as a result, is prone to sensing noise and other inaccurate data, leading to higher variations

ii.	For a higher ‘K’ value, the model is less sensitive to these fluctuations and as a result the end prediction is much more stable and results in smoother boundaries. Thereby, resulting in lower variance levels.


3)	Training Mean Squared Error (MSE):


i.	The error value in the training set will decrease for a higher ‘K’ value because the model is able to average over a large number of data points, thereby reducing the average error value.

ii.	A lower ‘K’ value will perfectly fit the training data, but when applied to another dataset, may result in higher error values because it essentially overfits the training data (as discussed above)


4)	Test Mean Squared Error (MSE):

  
i.	Similar to how Training MSE is observed, for an increasing value of ‘K’, we can see that the Test MSE Value will decrease. This is because, as mentioned above, the model averages out over a larger number of datapoints and as a result, the prediction is much more stable and smoother

ii.	However, upon continuing to increase the ‘K’ value will also result in the Test MSE will start to rise back up. This is for a very high value of ‘K’, the model will have very high bias and very low variance (i.e. underfits). This essentially means the model is over generalizing the dataset by averaging out too many neighbors.

iii.	Typically, the ideal ‘K’ value is found somewhere in between these 2 increasing values. It requires a degree of tuning to be found. 


5)	Irreducible Error:


i.	This error is basically an inherent characteristic of the dataset. It cannot be reduced by any model and is indicative of the permanent noise/imperfections in the data. Due to which, it remains constant regardless of any ‘K’ value.

**(e) Calculating the square of the bias**

*Bias is the term that measures how much the prediction varies from the tru functions output. Qw find that by finding the average difference between the predicted values and the actual values amd squaring them for all 15 degrees.* 

```python
bias_list = [] 

for degree in range(16):
    poly = PolynomialFeatures(degree=degree)
    X_poly = poly.fit_transform(X)
    model.fit(X_poly, y)
    y_pred = model.predict(X_poly)
    
    bias = np.mean(y_pred - y)
    bias = bias ** 2
    bias_list.append(bias)
    
for bias in bias_list:
    print(bias)
```

**(f). Finding variance**

*Variance is a measure of how much the predictions of the model vary. Here, we calculate how much it varies for each degree of polynomial. We calculate by computing the mean of the predicted values and then subsequently how much each prediction deviates from this mean. Squaring these differences will give variance*

```python
variance_list = []

for degree in range(16):
    poly = PolynomialFeatures(degree=degree)
    X_poly = poly.fit_transform(X)
    model.fit(X_poly, y)
    y_pred = model.predict(X_poly)
    
    mean_ypred = np.mean(y_pred)
    variance = np.mean((y_pred - mean_ypred)**2)
    variance_list.append(variance)

for variance in variance_list:
    print(variance)
```

**(c). Setting a randomseed of 123 and computing LOOCV errors from fitting the polynomials to the dataset.**

*Leave-One-Out Cross-Validation (LOOCV) works by seperating one data point for testing and training the model on the remaining n-1 datapoints (size of dataset is n). Then the process is repeated 'n' times where each time, each subsequent data-point is left out and the model is trained on the rest*

```python
np.random.seed(123)
```

```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_validate, LeaveOneOut, cross_val_score
```

```python
loo = LeaveOneOut() #this object is used to apply LOOCV when splitting the data during cross-validation
```

```python
loocv_errors = [] #list used to store the errors for each polynomial (1-7)
```

```python
for degree in range(1, 8):
    poly = PolynomialFeatures(degree) #creates the polynomial features
    X_poly = poly.fit_transform(x.reshape(-1, 1)) #transforms the values to correspoinding polynomials features

    model = LinearRegression() #fitting the linear regression model to each degree

    #cross-val-score is used to perform cross validation. here, because we are using LOOCV, we metion cv=obj 
    #scoring is used to specify the parameter used to evaluate the model, here its MSE
    mse_scores = -cross_val_score(model, X_poly, y, cv=loo, scoring='neg_mean_squared_error')
    loocv_error = np.mean(mse_scores)

    #the resultant mse_scores is an array of MSE values where each value represents the error of the dataset where one point is left out 
    loocv_errors.append(loocv_error) #computed error is added to the list 
```

*in the case if the cross_val_score function, we mention a negative sign because by default, cross_val_score is a function that correlates higher values with better performance. In our case, MSE is a loss-function. But sci-kit learn will invert our MSE into a negative MSE value to comply with the higher is better idea. This is just how sci-kit learn works. We aim to negate that effect inorder to get the accurate MSE valules.*

```python
import pandas as pd

loocv_errors_df = pd.DataFrame({
    'Degree': range(1, 8),
    'LOOCV Error': loocv_errors
})

# Print the DataFrame
print(loocv_errors_df)
```
